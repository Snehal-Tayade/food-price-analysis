```python
# Assuming 'Price' is the column name that represents restaurant prices
# Modify this to the correct column name in your dataset
data = pd.read_csv('zomato_dataset.csv')
high_rate_restaurant = data[(data['Dining_Rating'] > 4.5) & (data['Prices'] < 500)]  # Replace 500 with your threshold for low prices

# Display the top 10 results
print(high_rate_restaurant.head(10))
```

```python
# Define the price bins (you can adjust the ranges as needed)
price_bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

# Create a new column 'Price_Range' that categorizes the prices into bins
data['Price_Range'] = pd.cut(data['Prices'], bins=price_bins)

# Count how many food items fall into each price range
price_distribution = data['Price_Range'].value_counts().sort_index()

# Convert the result to a DataFrame for better formatting
price_distribution_df = pd.DataFrame(price_distribution).reset_index()

# Rename the columns for clarity
price_distribution_df.columns = ['Price Range', 'Frequency of Food Items']

# Display the result
print(price_distribution_df)
```

```python
# Group by 'Item_Name' and sum the 'Votes'
top_selling_items = data.groupby('Item_Name')['Votes'].sum().reset_index()

# Sort the items by total votes in descending order and get the top 10
top_10_selling_items = top_selling_items.sort_values(by='Votes', ascending=False).head(10)

# Display only the 'Item_Name' and 'Votes' columns
result = top_10_selling_items[['Item_Name', 'Votes']]

# Display the result without the index
print(result.to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt  # Import matplotlib
import seaborn as sns  
```

```python
# Load your dataset
data = pd.read_csv('zomato_dataset.csv')

average_price = data.groupby('Restaurant_Name')['Prices'].mean().reset_index()
top_5_restaurants = average_price.nlargest(5, 'Prices')

# Plotting
plt.figure(figsize=(12, 6))  # Adjust the size of the figure if needed
sns.barplot(x='Restaurant_Name', y='Prices', data=top_5_restaurants, palette='Set3')

# Customizing the plot
plt.title('Average prices for Restaurant')
plt.xlabel('Restaurant')
plt.ylabel('Average Price')

# Rotating x-axis labels for readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv('zomato_dataset.csv')

# Filter out restaurants with ratings available (you can adjust based on your dataset)
data_with_ratings = data.dropna(subset=['Dining_Rating'])

# Calculate the average prices and ratings for each restaurant
avg_price_rating = data_with_ratings.groupby('Restaurant_Name').agg({
    'Prices': 'mean',
    'Dining_Rating': 'mean'
}).reset_index()

# Select the top 15 restaurants based on average price or rating
top_5_restaurants = avg_price_rating.nlargest(5, 'Prices')

# Plotting
plt.figure(figsize=(12, 6))  # Adjust the figure size
sns.barplot(x='Restaurant_Name', y='Prices', hue='Dining_Rating', data=top_5_restaurants, palette='coolwarm')

# Customizing the plot
plt.title('Price vs Restaurant with Restaurant Rating')
plt.xlabel('Restaurant')
plt.ylabel('Average Price')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

```python
data = pd.read_csv('zomato_dataset.csv')
# Filter data for the top 15 restaurants based on the frequency of food items
top_5_restaurant_data = data[data['Restaurant_Name'].isin(top_5_restaurants['Restaurant_Name'])]

# Plotting the distribution of prices for the top 15 restaurants
plt.figure(figsize=(12, 6))  # Adjust the size of the figure
sns.histplot(data=top_5_restaurant_data, x='Restaurant_Name' , multiple='stack', bins=10)

# Customizing the plot
plt.title('Distribution of Prices for Top 5 Restaurants')
plt.xlabel('Restaurant_Name')
plt.ylabel('Frequency of Food Items')

# Display the plot
plt.show()


```

```python
from sklearn.cluster import KMeans

# select featur for clustering
x_cluster = data[['Dining Votes', 'Delivery_Votes', 'Votes', 'Prices']]

# fit KMeans model
kmeans_model = KMeans(n_clusters=5, random_state=42)
data['Cluster'] = kmeans_model.fit_predict(x_cluster)

# visualize the cluster
plt.figure(figsize=(10, 7))
sns.scatterplot(x = 'Dining Votes', y = 'Prices', hue='Cluster', data = data)
plt.xlabel('Dining Votes')
plt.ylabel('Prices')
plt.title('KMeans Clustering : Dining Votes vs. Prices')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Filter top 10 bestselling food items based on votes
top_bestsellers = data[data['Best_Seller'].notnull()].groupby('Item_Name')['Votes'].sum().nlargest(10)

# Step 2: Find the minimum price for each bestselling food item
min_prices = data[data['Item_Name'].isin(top_bestsellers.index)].groupby('Item_Name')['Prices'].min()

# Step 3: Plotting the top 10 bestselling items and their minimum prices with colorful bars using matplotlib
plt.figure(figsize=(10, 6))

# Generate a color map with different colors for each bar
colors = plt.cm.viridis(np.linspace(0, 1, len(top_bestsellers)))

# Create a horizontal bar plot with colors
plt.barh(top_bestsellers.index, min_prices.loc[top_bestsellers.index], color=colors)

# Labeling the axes
plt.xlabel('Minimum Price')
plt.ylabel('Bestselling Food Items')
plt.title('Top 10 Bestselling Food Items by Votes with Minimum Prices')

# Invert the y-axis to have the highest value at the top
plt.gca().invert_yaxis()

plt.show()

```
